3.9 (Calculating Time) In Exercise 2.11 , you wrote a script that calculated the number
of hours, minutes, and remaining seconds based on the number of seconds received
through user input. Reimplement your script to use a loop that in an iterative process
"picks off' the hours, minutes, and remaining seconds using the // and % operators.

```python
#Get total seconds from the user
total_seconds = int(input("Enter the total seconds: "))

hours = 0
minutes = 0
seconds = 0

while total_seconds > 0:
    # Determine how many hours
    if total_seconds >= 3600:
        hours = total_seconds // 3600
        total_seconds = total_seconds % 3600

    # Determine how many minutes
    elif total_seconds >= 60 :
        minutes = total_seconds // 60
        total_seconds = total_seconds % 60

    # Determine remaining seconds
    else:
        seconds = total_seconds
        total_seconds = 0

# Printing result
print("Hours: ", hours)
print("Minutes: ", minutes)
print("Seconds: ", seconds)
```

```python
# Initialization
total_does = 0
total_rabbits_born = 0
months = 0

# Sentinel-controlled loop
while True:
    # Input number of does and rabbits born
    does = int(input("Enter the number of does living in the farm (enter -1 to quit): "))
    
    # If sentinel value is entered, break the loop
    if does == -1:
        break
    
    rabbits_born = int(input("Enter the number of rabbits born in the month: "))

    # Update totals
    total_does += does
    total_rabbits_born += rabbits_born
    months += 1

# Calculate average
average = total_rabbits_born / total_does

# Output result
print("The combined average number of rabbits born per doe is ", average)
```

```python
n = int(input("Enter a non-negative integer: "))

sum_divi = 0

for i in range(1,n):
    if n % i == 0:
         sum_divi += i
        

if sum_divi == n:
     print(f"{n} is a perfect number.")
else:
    print(f"{n} is not a perfect number.")

```

```python
highest  = float("-inf")
second_highest = float("-inf")

for i in range (1,11):

    speed = int(input(f"Enter the speed runner {i} in m/sec: "))

    if speed > highest:
        second_highest = highest
        highest = speed
    elif speed > second_highest:
        second_highest = speed
print("\nThe fastest speed is: ", highest, "m/sec")
print("The second fastest speed is: ", second_highest, "m/sec")
```
